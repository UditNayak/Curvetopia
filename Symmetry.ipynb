```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load the image
image = cv2.imread('triangle.jpeg')

# Check if the image was successfully loaded
if image is None:
    print("Error: Image not found or unable to read.")
else:
    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply horizontal flipping
    flipped_horizontal = cv2.flip(gray, 1)

    # Draw a vertical Line to visualise symmetry
    img_symmetry = image.copy()
    cv2.line(img_symmetry, (gray.shape[1] // 2, 0), (gray.shape[1] // 2, gray.shape[0]), (0, 255, 0), 2)

    # Display the original image and the symmetry visualisation
    cv2_imshow(image)
    cv2_imshow(img_symmetry)

    # Apply vertical flipping
    flipped_vertical = cv2.flip(gray, 0)

```
